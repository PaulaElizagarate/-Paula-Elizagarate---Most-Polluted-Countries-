### Conexión con MySQL

Primero realizaremos la conexión:

```python
password= getpass("Contraseña de MySQL")
bbdd= "polluted_countries"
conexion= f"mysql+pymysql://root:{password}@localhost/{bbdd}"
engine= alch.create_engine(conexion)
```

### Inserción Datos en MySQL 

```python
for index, row in paisescont.iterrows():
    
    id_pais = f"""SELECT idpolluted_countries FROM polluted_countries 
                    WHERE `Country/Region` = "{row['Country/Region']}" """
    
       
    res = list(engine.execute(id_pais))
    
    if len(res) !=0:
        print(f"Este pais: {row['Country/Region']} ya existe en la BBDD")
       
    else:
        query= f"""INSERT INTO polluted_countries( idpolluted_countries, `Country/Region`, `2021`, `2020`, `2019`, `2018`, Population)
               VALUES
               ({index}, "{row['Country/Region']}", {row['2021']}, {row['2020']}, {row['2019']}, {row['2018']}, {row['Population']})
               """
        engine.execute(query)
```

```python
#valido

for index, row in ciudadescont.iterrows():
    
    country_id= list(engine.execute(f"""select idpolluted_countries from polluted_countries where `Country/Region`= '{row['country']}';"""))
    
    if len(country_id) == 0:
        print(f"No puedo insertar la fila {index}")
        
    else:
        country_id = list(country_id)[0][0]
        query= f"""INSERT INTO polluted_cities VALUES
               ({index}, "{row['City']}", {row['2021']}, {row['JAN(2021)']}, {row['FEB(2021)']}, {row['MAR(2021)']}, {row['APR(2021)']}, {row['MAY(2021)']}, {row['JUN(2021)']}, {row['JUL(2021)']} , {row['AUG(2021)']}, {row['SEP(2021)']}, {row['OCT(2021)']}, {row['NOV(2021)']}, {row['DEC(2021)']}, {row['2020']}, {row['2019']}, {row['2018']},{row['2017']}, "{row['country']}", {country_id})
               """
        engine.execute(query)

```

```python
#VALIDO
for index, row in nat.iterrows():
    
    count_id= list(engine.execute(f"""select idpolluted_countries from polluted_countries where `Country/Region`= '{row['country']}';"""))
    
    if len(count_id) == 0:
        print(f"No puedo insertar la fila {index}")
       
    else:
        count_id = list(count_id)[0][0]
        query= f"""INSERT INTO birth_rate
               VALUES
               ({index}, "{row['country']}", {row['year']}, {row['born']}, {row['male_born']}, {row['female_born']}, {row['birth_rate']}, {row['fertility_rate']}, {row['variation']}, {count_id})
               """
        engine.execute(query)
```

```python
#VALIDO

for index, row in costevida.iterrows():
    
    country_id= list(engine.execute(f"""select idpolluted_countries from polluted_countries where `Country/Region`= '{row['countries']}';"""))
    
    
    if len(country_id) == 0:
        
        print(f"No puedo insertar la fila {index}")
       
    else:
        country_id = list(country_id)[0][0]
        query= f"""INSERT INTO cost_living
               VALUES
               ({index}, "{row['countries']}", {row['cost_living_index']}, {row['year']}, {country_id})
               """
        engine.execute(query)
```
